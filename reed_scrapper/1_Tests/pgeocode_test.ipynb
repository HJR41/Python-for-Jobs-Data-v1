```python
import pgeocode
import requests
import pandas as pd
```

```python
country = "GB"


download_url = f"https://symerio.github.io/postal-codes-data/data/geonames/{country}.txt"
```

```python
column_names = [
    "country_code",
    "postal_code", 
    "place_name", 
    "country_name", 
    "country_abv", 
    "county_name", 
    "admin_1", 
    "district_name", 
    "ons_code", 
    "latitude", 
    "longitude", 
    "accuracy"
]

print(download_url) # postcodes innercode (second half of the postcode is always three letters long!!!!
```

```python
response = requests.get(download_url)

if response.status_code == 200:
    print(f'Response sucess {response.status_code}')
    df_geo = pd.read_csv(download_url, sep="\t", header=None, names=column_names, dtype=str)
    
else:
    print(f'Unsuccesful {response.status_code}')
```
